```
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np

print(' Changing Brightness Imagery ')
print('-----------------------------')
try:
    brightness = int(input('Insert Brightness point: '))
except ValueError:
    print('Error, not a number')

original= cv.imread('/content/drive/MyDrive/39daph-4.jpg')
brightness_image = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    for c in range(original.shape[2]):
      brightness_image[y,x,c] = np.clip(original[y,x,c]+ brightness, 0, 255)

final_frame = cv.hconcat((original, brightness_image))
cv2_imshow(final_frame)
```

```
grayavg = np.zeros(original.shape, original.dtype)
graylght = np.zeros(original.shape, original.dtype)
graylumin = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    grayavg[y,x] = np.array(np.sum(original[y,x])/3)
    graylght[y,x] = np.array((np.max(original[y,x])/2)+(np.min(original[y,x])/2))

    r,g,b = original[y,x]
    graylumin[y,x] = np.array((0.21*r)+(0.72*g)+(0.07*b))

print(original[1,1])
print('Grayscale_Average')
final_frame1 = cv.hconcat((original,grayavg))
cv2_imshow(final_frame1)
print('Grayscale_Lightness')
final_frame2 = cv.hconcat((original,graylght))
cv2_imshow(final_frame2)
print('Grayscale_Luminance')
final_frame3 = cv.hconcat((original,graylumin))
cv2_imshow(final_frame3)
```

```
cl = np.zeros(original.shape, original.dtype)

for y in range(original.shape[0]):
  for x in range(original.shape[1]):
    if original[y,x,2] >= 100 and original[y,x,1]<100:
      cl[y,x] = original[y,x]
    else:
      cl[y,x] = np.array(np.sum(original[y,x])/3)

final_frame = cv.hconcat((original, cl))
cv2_imshow(final_frame)
```
